```python
import random
import datetime

# Chatbot Header
print("=" * 60)
print("        🤖 AI Virtual Assistant Chatbot 🤖")
print("=" * 60)

# User Name Input
user_name = input("Enter your name: ")

print(f"\nWelcome {user_name}!")
print("Type 'help' to see available commands")
print("Type 'bye' to exit the chatbot")
print()

# Responses Database
responses = {

    "greetings": {
        "inputs": ["hello", "hi", "hey", "good morning", "good evening"],
        "responses": [
            f"Hello {user_name}! How can I assist you today?",
            f"Hi {user_name}! Welcome to the AI chatbot.",
            f"Hey {user_name}! Nice to meet you."
        ]
    },

    "about": {
        "inputs": ["who are you", "what is your name"],
        "responses": [
            "I am an AI Virtual Assistant developed using Python.",
            "You can call me PyBot Assistant."
        ]
    },

    "skills": {
        "inputs": ["what can you do", "help", "features"],
        "responses": [
            "I can answer questions and perform basic conversations.",
            "I can assist users using predefined AI responses.",
            "I can simulate a simple AI assistant."
        ]
    },

    "time": {
        "inputs": ["time", "current time"],
        "responses": []
    },

    "date": {
        "inputs": ["date", "today date"],
        "responses": []
    },

    "creator": {
        "inputs": ["who created you", "developer"],
        "responses": [
            f"I was developed by Teja using Python.",
            f"Teja created me as an AI internship project."
        ]
    },

    "bye": {
        "inputs": ["bye", "exit", "quit"],
        "responses": [
            f"Goodbye {user_name}! Have a wonderful day.",
            "See you later!",
            "Thank you for chatting with me."
        ]
    }
}

# Chatbot Function
def chatbot():

    while True:

        user_input = input(f"{user_name} : ").lower()

        matched = False

        # Response Matching
        for category in responses:

            for pattern in responses[category]["inputs"]:

                if pattern in user_input:

                    matched = True

                    # Time Feature
                    if category == "time":

                        current_time = datetime.datetime.now().strftime("%I:%M %p")

                        print("Bot :", "Current time is", current_time)

                    # Date Feature
                    elif category == "date":

                        current_date = datetime.datetime.now().strftime("%d-%m-%Y")

                        print("Bot :", "Today's date is", current_date)

                    else:

                        print("Bot :", random.choice(responses[category]["responses"]))

                    break

            if matched:
                break

        # Exit Condition
        if user_input in ["bye", "exit", "quit"]:

            print("Chatbot session ended successfully.")
            break

        # Unknown Input
        if not matched:

            print("Bot : Sorry, I didn't understand your request.")

# Run Chatbot
chatbot()
```
